### 01) WAP to create Calculator module which defines functions like add, sub,mul and div. 
### Create another .py file that uses the functions available in Calculator module.

```python
import calculator as calc

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ")

if operation == '+':
    result = calc.add(num1, num2)
elif operation == '-':
    result = calc.sub(num1, num2)
elif operation == '*':
    result = calc.mul(num1, num2)
elif operation == '/':
    result = calc.div(num1, num2)
else:
    result = "Invalid operation!"

print(f"Result: {result}")
```

### 02) WAP to pick a random character from a given String.

```python
import random

# random.choice()
my_string = "HELLO WORLD"
random_char = random.choice(my_string)
print(f"Random character: '{random_char}'")

# Method 2: Using random.randint()
string_length = len(my_string)
random_index = random.randint(0, string_length - 1)
random_char2 = my_string[random_index]
print(f"Random character (method 2): '{random_char2}'")
```

### 04) WAP to roll a dice in such a way that every time you get the same number.

### 05) WAP to generate 3 random integers between 100 and 999 which is divisible by 5.

```python
import random

count = 0
numbers = []

while count < 3:
    random_num = random.randint(100, 999)
    if random_num % 5 == 0:
        numbers.append(random_num)
        count += 1

for i, num in enumerate(numbers, 1):
    print(f"Number {i}: {num}")

# random.choice()

divisible_by_5 = [num for num in range(100, 1000) if num % 5 == 0]

random_picks = random.sample(divisible_by_5, 3)
print("\nRandom numbers divisible by 5 :" , random_picks)
```

### 06) WAP to generate 100 random lottery tickets and pick two lucky tickets from it and announce them as Winner and Runner up respectively.

```python
import random

print("Generating 100 lottery tickets...")
lottery_tickets = []

for i in range(100):
    ticket_id = f"TICKET-{i+1:03d}"
    lottery_tickets.append(ticket_id)

print(f"Generated {len(lottery_tickets)} tickets")
print(f"Sample tickets: {lottery_tickets[:5]}...{lottery_tickets[-5:]}")

print("\nLOTTERY DRAW RESULTS :")
print("=" * 30)

winners = random.sample(lottery_tickets, 2)
winner = winners[0]
runner_up = winners[1]

print(f"WINNER: {winner}")
print(f"RUNNER UP: {runner_up}")
```

### 08) Subtract a week (7 days) from a given date in Python.

```python
import datetime

try:
    year = int(input("Enter year (e.g., 2024): "))
    month = int(input("Enter month (1-12): "))
    day = int(input("Enter day (1-31): "))
    
    # Create date object
    user_date = datetime.date(year, month, day)
    print(f"\nYour date: {user_date.strftime('%A, %B %d, %Y')}")
    
    print(f"1 week ago: {(user_date - datetime.timedelta(days=7)).strftime('%A, %B %d, %Y')}")
    
except ValueError as e:
    print(f"Error: {e}")
    print("Please enter valid date values!")
    
```

### 09) WAP to Calculate number of days between two given dates.

```python
import datetime

try:
    print("Enter first date:")
    year1 = int(input("Year: "))
    month1 = int(input("Month (1-12): "))
    day1 = int(input("Day (1-31): "))
    
    print("\nEnter second date:")
    year2 = int(input("Year: "))
    month2 = int(input("Month (1-12): "))
    day2 = int(input("Day (1-31): "))
    
    first_date = datetime.date(year1, month1, day1)
    second_date = datetime.date(year2, month2, day2)
    
    date_diff = second_date - first_date
    
    print(f"First date: {first_date.strftime('%A, %B %d, %Y')}")
    print(f"Second date: {second_date.strftime('%A, %B %d, %Y')}")
    print(f"Difference: {abs(date_diff.days)} days")
        
except ValueError as e:
    print(f"Error: {e}")
    print("Please enter valid dates!")
```

### 10) WAP to Find the day of the week of a given date.(i.e. wether it is sunday/monday/tuesday/etc.)

```python
import datetime

try:
    print("Enter a date to find the day of the week:")
    user_year = int(input("Year: "))
    user_month = int(input("Month (1-12): "))
    user_day = int(input("Day (1-31): "))
    
    user_date = datetime.date(user_year, user_month, user_day)
    user_weekday = user_date.strftime("%A")
    
    print(f"Date: {user_date.strftime('%B %d, %Y')}")
    print(f"Day of the week: {user_weekday}")
    
    
except ValueError as e:
    print(f"Error: {e}")
    print("Please enter a valid date!")
```
